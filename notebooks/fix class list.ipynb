```python
import json
import os
```

```python
path = os.path.join('E:', 'age')
```

```python
os.listdir(path)
```

```python
with open(os.path.join(path, 'SAMPLE_age.json'), 'r') as fid:
    classes = json.load(fid)
```

```python
def depath(path):
    return os.path.split(path)[-1]

image_list = os.listdir(os.path.join(path, 'images'))

newclass = {}

for filename, label in classes.items():
    name = depath(filename)
    if name in image_list:
        newclass[name] = label
    else:
        print(f'{name} not found')
```

```python
newclass
```

```python
with open(os.path.join(path, 'classes.json'), 'w') as fid:
    json.dump(newclass, fid)
```
